## Getting started

Follow this gide to install the basic software and data requiered by this project

## Notebook

This is a Jupyter Notebook to show how to install all software required by this project

### Requirements
#### Basic programs
1. Python3
2. virtualenv3
3. C++

#### Bioinformatics programs
These programs have to be installed and included in the user PATH variable.

* STAR (https://github.com/alexdobin/STAR)
* SAMTools (http://www.htslib.org/)
* BAMTools (https://github.com/pezmaster31/bamtools)
* TPMCalculator (https://github.com/r78v10a07/TPMCalculator)

### Instalation of the Notebook

Execute this commands in a BASH terminal just the first time to get the notebook dependencies installed.

```
# Creating a virtual environment
$ virtualenv3 venv
$ source venv/bin/activate
$ pip install -r requirements/jupyter.txt
$ pwd
```

Edit the file in config/proj_env and change the PROJ_ROOT variable (first line)

### Downloading genomes files

* hg38 ftp://igenome:G3nom3s4u@ussd-ftp.illumina.com/Homo_sapiens/UCSC/hg38/Homo_sapiens_UCSC_hg38.tar.gz 
* hg19 ftp://igenome:G3nom3s4u@ussd-ftp.illumina.com/Homo_sapiens/UCSC/hg19/Homo_sapiens_UCSC_hg19.tar.gz

Uncompress this file and creates a data folder.

```
$ mkdir data
$ cd data
$ wget ftp://igenome:G3nom3s4u@ussd-ftp.illumina.com/Homo_sapiens/UCSC/hg38/Homo_sapiens_UCSC_hg38.tar.gz 
$ wget ftp://igenome:G3nom3s4u@ussd-ftp.illumina.com/Homo_sapiens/UCSC/hg19/Homo_sapiens_UCSC_hg19.tar.gz
$ tar xzfv Homo_sapiens_UCSC_hg38.tar.gz
$ tar xzfv Homo_sapiens_UCSC_hg19.tar.gz
$ mkdir hg19 hg38
$ mv Homo_sapiens/UCSC/hg38/Annotation/Genes/genes.gtf hg38/
$ mv Homo_sapiens/UCSC/hg38/Sequence/WholeGenomeFasta/genome.fa hg38/
$ mv Homo_sapiens/UCSC/hg19/Annotation/Genes/genes.gtf hg19/
$ mv Homo_sapiens/UCSC/hg19/Sequence/WholeGenomeFasta/genome.fa hg19/

```

### Instalation of software 

#### Compiling the STAR

```
$ cd src/
$ git clone https://github.com/alexdobin/STAR
$ cd STAR/source
$ make STAR
$ cd ../../../bin/
$ ln -s ../src/STAR/source/STAR STAR
$ cd ../
$ cd data/hg19/
$ mkdir STAR
$ cd STAR
$ ln -s ../genome.fa genome.fa
$ ln -s ../genes.gtf genes.gtf
$ ../../../bin/STAR --runThreadN 16 --runMode genomeGenerate --genomeDir . --genomeFastaFiles genome.fa --sjdbGTFfile ./genes.gtf
$ cd ../../../
```

#### Compiling the BAMTools

```
$ cd src/
$ git clone https://github.com/pezmaster31/bamtools
$ cd bamtools
$ mkdir build
$ cd build
$ cmake ..
$ make
$ cd ../../../bin/
$ ln -s ../src/bamtools/bin/bamtools bamtools
$ cd ../
$ source config/proj_env
```

#### Compiling TPMCalculator

```
$ cd src/
$ git clone https://github.com/r78v10a07/TPMCalculator
$ cd TPMCalculator
$ make 
$ cd ../../bin/
$ ln -s ../src/TPMCalculator/bin/TPMCalculator TPMCalculator
$ cd ../
```

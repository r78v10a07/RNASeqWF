## Retriving data from the BioProject ID using fastq-dump

```python
os.chdir(data_dir)
handle = Entrez.esearch(db="sra", term=os.environ['DATASET'], retmax=50)
sra_id = Entrez.read(handle)
handle.close()
tuples = []
!rm -rf commands
for id in sra_id['IdList']:
    handle = Entrez.efetch(db="sra", id=id, rettype="full", retmode="full")
    sra = xmltodict.parse(handle.read())
    handle.close()  
    sra_data = sra['EXPERIMENT_PACKAGE_SET']['EXPERIMENT_PACKAGE']
    alias = '' if 'SAMPLE' not in sra_data else sra_data['SAMPLE']['@alias']
    accession = '' if 'SAMPLE' not in sra_data else sra_data['SAMPLE']['@accession']
    if type(sra_data['RUN_SET']['RUN']) is list: 
        for run in sra_data['RUN_SET']['RUN']:
            f = run['@accession']
            tuples.append([alias, accession, f])
            !echo "/gfs/progs/sratoolkit.2.9.2-centos_linux64/bin/fastq-dump {f} --gzip" >> commands
    elif type(sra_data['RUN_SET']['RUN']) is collections.OrderedDict:
        f = sra_data['RUN_SET']['RUN']['@accession']
        tuples.append([alias, accession, f])
        !echo "/gfs/progs/sratoolkit.2.9.2-centos_linux64/bin/fastq-dump {f} --gzip" >> commands
f = os.path.join(os.environ['DATA'], os.environ['DATASET'], 'samples.tsv')
df = pandas.DataFrame(tuples, columns=['ALIAS', 'ACCESSION', 'RUN'])
df.to_csv(f, sep='\t', index=False, na_rep='0')
display(df)
!cat commands | parallel -j 16
```

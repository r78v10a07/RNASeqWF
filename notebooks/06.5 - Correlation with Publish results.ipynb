```python
xls_file = os.path.join(os.environ['DATA'], os.environ['DATASET'], '12859_2016_1457_MOESM2_ESM.xlsx')
xls_df = pandas.read_excel(xls_file, sheet_name=None, header=None);
for k in xls_df:
    if not k.startswith('int-'):
        xls_df[k].columns = xls_df[k].iloc[0]
        xls_df[k] = xls_df[k].reindex(xls_df[k].index.drop(0))
```

```python
gene_gtf_file = "/gfs/data/genomes/Gencode/Gencode_human/GRCh37/release_19/gencode.v19.annotation_genes.gtf"
gene_gtf = pandas.read_csv(gene_gtf_file, sep='\t',header=None)
gene_gtf = gene_gtf[[9,10]]
gene_gtf = gene_gtf.rename(index=str, columns={9:'Gene_Id', 10:'gene_name'})
gene_gtf.head()
```

```python
result_dir = os.path.join(os.environ['RESULTS'], os.environ['DATASET'], 'differential_expression')
os.chdir(result_dir)
```

```python
how='inner'

dga_results = pandas.read_csv('samseq.csv')
print('DGA Samseq results: ' + str(len(dga_results)))

dga_edgeR = pandas.read_csv('edgeR.csv')
print('DGA EdgeR results: ' + str(len(dga_edgeR)))
dga_results = dga_results.merge(dga_edgeR, left_on='Gene Name', right_on='Gene_Id', how=how)
dga_results['Gene_Id'].fillna(dga_results['Gene Name'],inplace=True)

dga_deseq2 = pandas.read_csv('deseq2.csv')
print('DGA Deseq2 results: ' + str(len(dga_deseq2)))
dga_results = dga_results.merge(dga_deseq2, on='Gene_Id', how=how)

dga_mwu = pandas.read_csv('mwu.csv')
print('DGA MWU results: ' + str(len(dga_mwu)))
dga_results = dga_results.merge(dga_mwu, on='Gene_Id', how=how)

dga_results = dga_results.merge(gene_gtf, on='Gene_Id', how='inner')
dga_results = dga_results.sort_values(by='gene_name')
print('DGA intercepted results: ' + str(len(dga_results)))

a = []
for k in xls_df:
    if k.startswith('int-'):
        c = len(xls_df[k].merge(dga_results, left_on=0, right_on='gene_name', how='inner'))   
    else:
        c = len(xls_df[k].merge(dga_results, left_on='Gene Name', right_on='gene_name', how='inner'))
    r = c/(len(xls_df[k]) - 50)
    p = c/len(dga_results)
    a.append([k,len(xls_df[k]), len(dga_results), c, r, p])
interception = pandas.DataFrame(a, columns=['Reference', 'Genes in Ref', 'Genes', 'Common',  'Recall', 'Precision'])
average = interception.mean(axis=0)
display(interception)
interception.to_csv('intercept_corr.csv', sep='\t',index=None)
print("No. Genes: " + str(average['Common']) + " Recall: " + str(average['Recall']) + ' Precision: ' + str(average['Precision']))
```

```python
how='outer'

dga_results = pandas.read_csv('samseq.csv')
print('DGA Samseq results: ' + str(len(dga_results)))

dga_edgeR = pandas.read_csv('edgeR.csv')
print('DGA EdgeR results: ' + str(len(dga_edgeR)))
dga_results = dga_results.merge(dga_edgeR, left_on='Gene Name', right_on='Gene_Id', how=how)
dga_results['Gene_Id'].fillna(dga_results['Gene Name'],inplace=True)

dga_deseq2 = pandas.read_csv('deseq2.csv')
print('DGA Deseq2 results: ' + str(len(dga_deseq2)))
dga_results = dga_results.merge(dga_deseq2, on='Gene_Id', how=how)

dga_mwu = pandas.read_csv('mwu.csv')
print('DGA MWU results: ' + str(len(dga_mwu)))
dga_results = dga_results.merge(dga_mwu, on='Gene_Id', how=how)

dga_results = dga_results.merge(gene_gtf, on='Gene_Id', how='inner')
dga_results = dga_results.sort_values(by='gene_name')
print('DGA intercepted results: ' + str(len(dga_results)))

a = []
for k in xls_df:
    if k.startswith('int-'):
        c = len(xls_df[k].merge(dga_results, left_on=0, right_on='gene_name', how='inner'))   
    else:
        c = len(xls_df[k].merge(dga_results, left_on='Gene Name', right_on='gene_name', how='inner'))
    r = c/(len(xls_df[k]) - 50)
    p = c/len(dga_results)
    a.append([k,len(xls_df[k]), len(dga_results), c, r, p])
union = pandas.DataFrame(a, columns=['Reference', 'Genes in Ref', 'Genes', 'Common',  'Recall', 'Precision'])
average = union.mean(axis=0)
display(union)
union.to_csv('union_corr.csv', sep='\t',index=None)
print("No. Genes: " + str(average['Common']) + " Recall: " + str(average['Recall']) + ' Precision: ' + str(average['Precision']))
```

## Plotting 

```python
union = union.drop(columns=['Genes', 'Common'])
interception = interception.drop(columns=['Genes in Ref','Genes', 'Common'])

edgeR = pandas.read_csv('edgeR_corr.csv', sep='\t')
edgeR = edgeR.drop(columns=['Genes in Ref','Genes', 'Common'])

deseq = pandas.read_csv('deseq2_corr.csv', sep='\t')
deseq = deseq.drop(columns=['Genes in Ref','Genes', 'Common'])

samseq = pandas.read_csv('samseq_corr.csv', sep='\t')
samseq = samseq.drop(columns=['Genes in Ref','Genes', 'Common'])

mwu = pandas.read_csv('mwu_corr.csv', sep='\t')
mwu = mwu.drop(columns=['Genes in Ref','Genes', 'Common'])
```

```python
col = 'Recall'
df = union[['Reference',col]].merge(interception[['Reference',col]], on='Reference')
df = df.rename(columns={col + '_x': 'Union', col + '_y': 'Interception'})

df = df.merge(edgeR[['Reference',col]], on='Reference')
df = df.rename(columns={col: 'EdgeR'})

df = df.merge(deseq[['Reference',col]], on='Reference')
df = df.rename(columns={col: 'Deseq'})

df = df.merge(samseq[['Reference',col]], on='Reference')
df = df.rename(columns={col: 'SAMseq'})

df = df.merge(mwu[['Reference',col]], on='Reference')
df = df.rename(columns={col: 'MWU'})

display(df)
df.to_csv('recall.csv', index=None, sep='\t')
df.plot.box(title='Recall')
df.plot.barh(figsize=(12, 10), title='Recall').set_yticklabels(df['Reference'], rotation=0);
```

```python
col = 'Precision'
df = union[['Reference',col]].merge(interception[['Reference',col]], on='Reference')
df = df.rename(columns={col + '_x': 'Union', col + '_y': 'Interception'})

df = df.merge(edgeR[['Reference',col]], on='Reference')
df = df.rename(columns={col: 'EdgeR'})

df = df.merge(deseq[['Reference',col]], on='Reference')
df = df.rename(columns={col: 'Deseq'})

df = df.merge(samseq[['Reference',col]], on='Reference')
df = df.rename(columns={col: 'SAMseq'})

df = df.merge(mwu[['Reference',col]], on='Reference')
df = df.rename(columns={col: 'MWU'})

display(df)
df.to_csv('precision.csv', index=None, sep='\t')
df.plot.box(title='Precision')
df.plot.barh(figsize=(12, 10), title='Precision').set_yticklabels(df['Reference'], rotation=0);
```
